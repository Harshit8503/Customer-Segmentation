```python
from sklearn.cluster import KMeans
```

```python
error=[]
df1=df[["Annual Income (k$)","Spending Score (1-100)"]]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df1)
    error.append(kmeans.inertia_)
```

```python
km=KMeans(n_clusters=5)
y=km.fit_predict(df1)
df1["Label"]=y
```

```python
sns.scatterplot(data=df1,x="Annual Income (k$)",y="Spending Score (1-100)",hue="Label")
```

```python
error=[]
df2=df[["Annual Income (k$)","Spending Score (1-100)","Age"]]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df2)
    error.append(kmeans.inertia_)
```

```python
plt.plot(np.linspace(1,10,10),error)
```
